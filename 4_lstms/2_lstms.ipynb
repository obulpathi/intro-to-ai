```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt


F = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 21.2, 8.6, -5.8]
C = [2.05,  13.27, 14.5, 28.0, 11.0,  8.0,  3.0, -4.0,  -6.0, -13.0, -21.0]


def f(w, b):
    CE = [w * f + b for f in F] 
    loss = 0.0
    for ce, c in zip(CE, C):
        loss = loss + (ce - c)**2
    loss = loss / len(F)
    return loss


w = np.linspace(-2.5, 2.5, 100)
b = np.linspace(-50, 10, 100)

W, B = np.meshgrid(w, b)
L = f(W, B)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(W, B, L, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.view_init(90, 0)
ax.set_zlabel('l');
```
